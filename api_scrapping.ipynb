```python
import json
import requests
import pandas as pd
```

```python

url = "https://data.police.uk/api/forces"
params = {}
response = requests.get(url)
print(response.request.url)
result = response.json()
response
```

```python
dateas = pd.DataFrame(result)
dateas = dateas.drop(["name"], axis = 1)
forces = dateas["id"].tolist()
forces.remove("metropolitan")
forces = ['city-of-london','west-midlands','greater-manchester','west-yorkshire','kent','hampshire','essex',
               'lancashire','merseyside','south-yorkshire','devon-and-cornwall','surrey','hertfordshire',
               'nottinghamshire','staffordshire','north-yorkshire','lincolnshire','cheshire','leicestershire'
               'derbyshire','northumbria']
forces
```

```python
url = "https://data.police.uk/api/crimes-street-dates"
response1 = requests.get(url)
result1 = response1.json()
dates = pd.DataFrame(result1)
dates = dates.drop(["stop-and-search"], axis = 1)
dates = dates["date"].tolist()
dates = dates[::-1]
dates
```

```python
#ESTA ES
data_temp = pd.DataFrame()
datafirst = pd.DataFrame()
for i in range(5,8):
    for j in dates:
        params = {"force": forces[i],
                     "date": j}
        response1 = requests.get("https://data.police.uk/api/stops-force",params)
        result1 = response1.json()
        temp = pd.DataFrame(result1)
        data_temp = data_temp.append(temp)
    data_temp["force"] = forces[i]
    datafirst = datafirst.append(data_temp, ignore_index=True)
```

```python
#SECOND CHANCE
data_temp = pd.DataFrame()
datasecond = pd.DataFrame()
for i in range(8, len(forces)):
    for j in dates:
        params = {"force": forces[i],
                     "date": j}
        response1 = requests.get("https://data.police.uk/api/stops-force",params)
        result1 = response1.json()
        temp = pd.DataFrame(result1)
        data_temp = data_temp.append(temp)
    data_temp["force"] = forces[i]
    datasecond = datasecond.append(data_temp, ignore_index=True)
```
